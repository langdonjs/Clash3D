```python
import ultralytics
from ultralytics import YOLO
import torch
import os
import cv2

EPOCHS = 1
IMG_SIZE = 896

os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
ultralytics.utils.checks.collect_system_info()
```

```python
model = YOLO('yolo26n.pt')
model.to('cuda' if torch.cuda.is_available() else 'cpu')
results = model.train(data = './clash_royacado_tiny/data.yaml', epochs = EPOCHS, imgsz = IMG_SIZE, batch = 8, project ="./models",
                    name="clash_yolo_v1", rect=True, mosaic=0.0, scale=0.0, degrees=0.0, perspective=0.0,
                     amp=True,exist_ok=True,save_period=5)
```
